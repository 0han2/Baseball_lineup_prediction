# 컬럼 정리

```python
df1 = pd.read_csv('NC_defense_detail_2020.csv')
df2 = pd.read_csv('doosan_defense_detail_2020.csv')
df3 = pd.read_csv('Samsung_defense_detail_2020.csv')
df4 = pd.read_csv('LG_defense_detail_2020.csv')
df5 = pd.read_csv('kiwooom_defense_detail_2020.csv')

df6 = pd.read_csv('lotte_defense_detail_2020.csv')
df7 = pd.read_csv('SSG_defense_detail_2020.csv')
df8 = pd.read_csv('Hanwha_defense_detail_2020.csv')
df9 = pd.read_csv('KIA_defense_detail_2020.csv')
df10 = pd.read_csv('KT_defense_detail_2020.csv')
```

```python
# df1에서 'rank' 열이 'rank'인 행 제거
df1= df1[df1['Rank'] != 'Rank']
df2= df2[df2['Rank'] != 'Rank']
df3= df3[df3['Rank'] != 'Rank']
df4= df4[df4['Rank'] != 'Rank']
df5= df5[df5['Rank'] != 'Rank']
df6= df6[df6['Rank'] != 'Rank']
df7= df7[df7['Rank'] != 'Rank']
df8= df8[df8['Rank'] != 'Rank']
df9= df9[df9['Rank'] != 'Rank']
df10= df10[df10['Rank'] != 'Rank']

# 컬럼명 한글이름으로 바꾸기
columns_to_keep = ['Name', 'Team', 'Sort▼']
df1 = df1[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df2 = df2[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df3 = df3[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df4 = df4[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df5 = df5[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df6 = df6[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df7 = df7[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df8 = df8[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df9 = df9[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})
df10 = df10[columns_to_keep].rename(columns={'Sort▼': '수비 승리 기여도', 'Name': '선수명', 'Team': '포지션'})

# 포지션 컬럼에서 앞 세 글자 제외
df1['포지션'] = df1['포지션'].str[3:]
df2['포지션'] = df2['포지션'].str[3:]
df3['포지션'] = df3['포지션'].str[3:]
df4['포지션'] = df4['포지션'].str[3:]
df5['포지션'] = df5['포지션'].str[3:]
df6['포지션'] = df6['포지션'].str[3:]
df7['포지션'] = df7['포지션'].str[3:]
df8['포지션'] = df8['포지션'].str[3:]
df9['포지션'] = df9['포지션'].str[3:]
df10['포지션'] = df10['포지션'].str[3:]

position_mapping = {
    '1B': '1루수',    '2B': '2루수',    '3B': '3루수',    'SS': '유격수',
    'C': '포수',    'RF': '우익수',    'CF': '중견수',    'LF': '좌익수'
}

# 포지션 컬럼 값 변환
df1['포지션'] = df1['포지션'].map(position_mapping)
df2['포지션'] = df2['포지션'].map(position_mapping)
df3['포지션'] = df3['포지션'].map(position_mapping)
df4['포지션'] = df4['포지션'].map(position_mapping)
df5['포지션'] = df5['포지션'].map(position_mapping)
df6['포지션'] = df6['포지션'].map(position_mapping)
df7['포지션'] = df7['포지션'].map(position_mapping)
df8['포지션'] = df8['포지션'].map(position_mapping)
df9['포지션'] = df9['포지션'].map(position_mapping)
df10['포지션'] = df10['포지션'].map(position_mapping)

df1['년도'] = '2020'
df2['년도'] = '2020'
df3['년도'] = '2020'
df4['년도'] = '2020'
df5['년도'] = '2020'
df6['년도'] = '2020'
df7['년도'] = '2020'
df8['년도'] = '2020'
df9['년도'] = '2020'
df10['년도'] = '2020'
```
